```python
from sklearn import datasets
iris = datasets.load_iris()
# list(iris.keys())

a = iris['data']
print(a.shape)
b = iris['target']
print(b.shape)
b
```

```python
# simple logistic regression model on Virginica using one feature, petal width

import numpy as np
from matplotlib import pyplot as plt

X = a[:,3:] # petal width
y = (b == 2).astype(int) # virginica is 2, and y is boolean w/o astype
#print(y)

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,y)

X_new = np.linspace(0,3,1000).reshape(-1,1) # reshape list into a 2D array with a shape (n, 1)
y_proba = log_reg.predict_proba(X_new)
# print(y_proba.shape)
# print(y_proba)

plt.plot(X_new,y_proba[:,0],'b--',label='not virginica')
plt.plot(X_new,y_proba[:,1],'r-',label='virginica')
plt.xlabel('petal width')
plt.ylabel('proba')
plt.legend()

print(log_reg.predict([[1.7],[1.5]]))
```

```python
# multi-class softmax

X = iris['data'][:,2:]
y = iris['target']

# must also specify a solver that support Softmax regression
# it also provides regularization with hyperparameter C

softmax_reg = LogisticRegression(multi_class='multinomial',solver='lbfgs',C =10)
softmax_reg.fit(X,y)

softmax_reg.predict([[5,2]])
softmax_reg.predict_proba([[5,2]])
# [5,2] has a probability of 0.94 as virginica, while 5.7% as versicolor
```
